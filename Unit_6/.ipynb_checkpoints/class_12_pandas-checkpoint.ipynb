```python
#importing an excel file
data = pd.read_csv('products-2.csv', sep='\t')
```

```python
# first five rows

data.head()
```

```python
# last five rows

data.tail()
```

```python
#columns

data.columns
```

```python
#summary statistics
data.describe()
```

```python
# what items have list prices less than $500: how to write a conditon

data[data['ListPrice'] > 3000]
```

```python
#display: list price, standard cost, , weight and color where price > 3000
# use [[]] -> to get data
data[data['ListPrice'] > 3000][['ListPrice','Weight','Color','StandardCost']]
```

```python
#all products that are red
data[data['Color'] == 'Red']
```

```python
#products silver and list price>500 -> data[(cond.1) & (cond.2)]
data[(data['Color'] == 'Silver') & (data['ListPrice']>500)]
```

```python
#add new column to 'to_reorder' = current stock - reorder point

data['to_reorder'] = data['SafetyStockLevel'] - data['ReorderPoint']
```

```python
data
```

```python
#how many different colors in data set: value.count()


data['Color'].value_counts()
```

```python
#how do we know how many unique values are in a col

data['Color'].unique()
```
